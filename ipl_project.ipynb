```python
#We have to find which team win more times
match_Wins = df['match_winner'].value_counts()
sns.barplot(x = match_Wins.values , y = match_Wins.index , palette='viridis')
plt.title('Most match Winner')
```

```python
df.head()
```

```python
#Find how many times toss discion for field and count
sns.countplot(x = df['toss_decision'] , palette='rainbow')
plt.title("Toss descion trend")
```

```python
# find the percentge of toss winner lies beetween wining 
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

```python
# In this we have to find team how many team is win by runs or wicket
df.head()
sns.countplot(x=df['won_by'])
plt.title('Win By')
```

**Key Player Performance**

```python
# 1 most Player of the match
df.head()

count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x = count.values , y = count.index , palette='rainbow')
plt.title("TOp 10 Player")
```

```python
# 2 Top scorers
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
```

```python
sns.barplot(x = high.index , y = high.values , palette='rainbow')
```

```python
high.plot(kind = 'barh')
```

```python
# 10 best Bowling figuers
df['highest_wicket'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wicket'] = df['highest_wicket'].astype(int)
high = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
sns.barplot(x = high.values , y = high.index , palette='rainbow')
high
```

Venue Analysis



```python
# Most Matches Played by venue
venue = df['venue'].value_counts()
sns.barplot(x = venue.values , y = venue.index, palette='rainbow')
```

```python
# Who won the highest margin by runs
df[df['won_by'] == 'Runs'].sort_values(by = 'margin' , ascending=False)[['margin' , 'match_winner']].head(1)
```

```python
 # which player had the highest individual score
# df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(1)
df[df['highscore'] == df['highscore'].max()][['top_scorer' , 'highscore']]
```

```python
#which player is the most best bowling figure
df[df['highest_wicket'] == df['highest_wicket'].max()][['best_bowling' ,'best_bowling_figure' ]]
```
